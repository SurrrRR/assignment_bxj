```python
import pandas as pd

qinghefang=[]

for i in range(1,5):
    page = requests.get('https://travel.qunar.com/p-cs300195-hangzhou-jingdian-1-'+str(i)+'?lc=1&location=%2525E6%2525B8%252585%2525E6%2525B2%2525B3%2525E5%25259D%25258A%2525E8%2525A1%252597&lat=30.239933&lng=120.169201&pid=714820&distance=1000') 
    print("\n***** Page " + str(i) +" in action *****")
    soup = BeautifulSoup(page.content,'html.parser')
    
    for j in soup.find_all("div", attrs={"class":"ct"}):
        page_no = str(i)
        name=j.a.get_text()
        #en_title=j.find("span",attrs={"class":"cn_tit"}).span.get_text()
        
        url = j.a.get('href')
        
        des=j.find("div",attrs={"class":"desbox"}).get_text()
        #print(type(des))
        #print(des)

        
        qinghefang.append({ "spot_name":name,"website":url,"description":des})

#print(names)
#print(urls)
#print(descriptions)

print(qinghefang)            
                
         
df=pd.DataFrame(qinghefang)
df
df.to_csv("around_qinghefang.csv", sep='\t', encoding='utf-8')
df


```
